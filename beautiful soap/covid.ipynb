```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url=r"https://www.worldometers.info/coronavirus/"
responce=requests.get(url)
responce.status_code
```

```python
soup=BeautifulSoup(responce.content,'lxml')
```

```python
soup.table.thead.findAll('th')[1:][0].text.replace(',','_')
```

```python
coloums=[]
for col in soup.table.tbody.findAll('th')[1:]:
    coloums.append(re.sub(
        r'[,\s/]','_',col.text
    ))
```

```python
coloums
```

```python
#collecting Data
data=[]
for row in soup.table.tbody.findAll('tr')[8:]:
    temp=[]
    for td in row.findAll('td'):
        temp.append(td.text)
    print(temp) 
    data.append(temp)
```

```python
data
```

```python
#Saving Data in csv flie
import csv
with open("Covid_information.csv", mode="w", newline='') as f:
    writer=csv.writer(f, delimetre=)
    writer.writerow(coloums)
    
```
